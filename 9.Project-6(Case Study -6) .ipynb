# Convertin into date time data type/parse_Dates

# Here the published_timestamp is in object formate but we neet to convert it into date time while importing

```python
df=pd.read_csv('udemy_courses.csv',parse_dates=["published_timestamp"])
```

```python
 df.info()
```

```python
df.head(4)
```

```python
#Which course has more number of lectures
df.groupby('is_paid').agg({'num_lectures':'mean'})
```

```python
#Which course has higer number of subscribers (Free or paid)
```

```python
df.groupby("is_paid")["num_subscribers"].mean()
```

```python
#Which level has the highest number of subscribers
```

```python
df.groupby('level')["num_subscribers"].mean()
```

```python
df["num_subscribers"].dtypes
```

```python
#10 most polular courses as per subscribers
```

```python
df.sort_values(by="num_subscribers",ascending=False).head(10)
```

```python
df["course_title"][df["num_reviews"]==df["num_reviews"].max()]
```

```python
df.groupby("is_paid")["num_reviews"].mean()
```

```python
#display 10 most popular courses of python as number of subscribers

df[df['course_title'].str.contains("python",case=False)].sort_values(by="num_subscribers",ascending=False).head(10)
```
